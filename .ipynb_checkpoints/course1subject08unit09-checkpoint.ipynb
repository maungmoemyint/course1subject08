```python
import seaborn as sns
```

```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

### Iris data

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

### Simple Plot

```python
sns.histplot(iris["sepal_length"], kde=True, stat="density")
plt.show()
```

### Regression plot

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)
plt.show()
```

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)

plt.title("Relationship of Sepal and Petal length of Iris Flowers") # Add plot title
plt.ylabel("Petal Length") # Adjust the label of the y-axis
plt.xlabel("Sepal Length") # Adjust the label of the x-axis
plt.ylim(0.5, 9) # Adjust the limits of the y-axis
plt.xlim(4, 8) # Adjust the limits of the x-axis

plt.show()
```

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)

plt.title("Relationship of Sepal and Petal length of Iris Flowers") # Add plot title
plt.ylabel("Petal Length") # Adjust the label of the y-axis
plt.xlabel("Sepal Length") # Adjust the label of the x-axis
plt.ylim(0.5, 9) # Adjust the limits of the y-axis
plt.xlim(4, 8) # Adjust the limits of the x-axis
sns.set_style("darkgrid")
plt.show()
```

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)

plt.title("Relationship of Sepal and Petal length of Iris Flowers") # Add plot title
plt.ylabel("Petal Length") # Adjust the label of the y-axis
plt.xlabel("Sepal Length") # Adjust the label of the x-axis
plt.ylim(0.5, 9) # Adjust the limits of the y-axis
plt.xlim(4, 8) # Adjust the limits of the x-axis
sns.set_style("ticks") #Execute it twice to see the ticks default format
plt.show()
```

### Pair Plots

```python
sns.pairplot(iris, height=1.5)
plt.show()
```

```python
sns.pairplot(iris, hue="species", corner=True, height=1.5)
plt.show()
```

```python
sns.pairplot(iris, x_vars=["sepal_length", "sepal_width"],
                   y_vars=["petal_length"],
                   hue="species",
                   height=3)
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 1, # sets the number of rows in the subplot grid
ncols = 2, # sets the number of columns in the subplot grid
figsize = (8, 3), # sets the width and height of the figure (in inches)
sharey=True) # let the two plots share the same y-axis

# the left-hand side scatter plot
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue='species',
ax=axes[0]) # draw the plot on the first axes

plt.show()
```

```python
fig, axes = plt.subplots(nrows = 1, # sets the number of rows in the subplot grid
ncols = 2, # sets the number of columns in the subplot grid
figsize = (8, 3), # sets the width and height of the figure (in inches)
sharey=True) # let the two plots share the same y-axis

# the left-hand side scatter plot
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue='species',
ax=axes[0]) # draw the plot on the first axes

axes[0].set_title('petal length vs sepal length')
axes[0].legend(loc='upper left') # set the legend location
axes[0].set_xlabel('Sepal Length')


plt.show()
```

```python
fig, axes = plt.subplots(nrows = 1, # sets the number of rows in the subplot grid
ncols = 2, # sets the number of columns in the subplot grid
figsize = (8, 3), # sets the width and height of the figure (in inches)
sharey=True) # let the two plots share the same y-axis

# the left-hand side scatter plot
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue='species',
ax=axes[0]) # draw the plot on the first axes

axes[0].set_title('petal length vs sepal length')
axes[0].legend(loc='upper left') # set the legend location
axes[0].set_xlabel('Sepal Length')


# the right-hand side scatter plot
sns.scatterplot(data=iris, x="sepal_width", y="petal_length", hue='species',
ax=axes[1], # draw the plot on the second axes
legend=False) # turn of the legend

plt.show()
```

```python
fig, axes = plt.subplots(nrows = 1, # sets the number of rows in the subplot grid
ncols = 2, # sets the number of columns in the subplot grid
figsize = (8, 3), # sets the width and height of the figure (in inches)
sharey=True) # let the two plots share the same y-axis

# the left-hand side scatter plot
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue='species',
ax=axes[0]) # draw the plot on the first axes

axes[0].set_title('petal length vs sepal length')
axes[0].legend(loc='upper left') # set the legend location
axes[0].set_xlabel('Sepal Length')


# the right-hand side scatter plot
sns.scatterplot(data=iris, x="sepal_width", y="petal_length", hue='species',
ax=axes[1], # draw the plot on the second axes
legend=False) # turn of the legend

axes[1].set_title('petal length vs sepal width')
axes[1].set_xlabel('Sepal Width')
fig.tight_layout()
plt.show()
```

```python
### color palette in seaborn

sns.pairplot(
    iris,
    x_vars=["sepal_length"],
    y_vars=["petal_length"],
    height=3.5,
    hue="species",
    palette="pastel",
)
plt.show()
```

```python
### Custom color platte

# define a custom palette
my_palette = ["#95a5a6", "#e74c3c", "#34495e"]
sns.pairplot(
    iris,
    x_vars=["sepal_length"],
    y_vars=["petal_length"],
    height=3.5,
    hue="species",
    palette=my_palette, # <-- see here, custom palette
)
plt.show()
```
