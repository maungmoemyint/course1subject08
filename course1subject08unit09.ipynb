```python
import seaborn as sns
```

```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

### Iris data

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

### Simple Plot

```python
sns.histplot(iris["sepal_length"], kde=True, stat="density")
plt.show()
```

### Regression plot

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)
plt.show()
```

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)

plt.title("Relationship of Sepal and Petal length of Iris Flowers") # Add plot title
plt.ylabel("Petal Length") # Adjust the label of the y-axis
plt.xlabel("Sepal Length") # Adjust the label of the x-axis
plt.ylim(0.5, 9) # Adjust the limits of the y-axis
plt.xlim(4, 8) # Adjust the limits of the x-axis

plt.show()
```

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)

plt.title("Relationship of Sepal and Petal length of Iris Flowers") # Add plot title
plt.ylabel("Petal Length") # Adjust the label of the y-axis
plt.xlabel("Sepal Length") # Adjust the label of the x-axis
plt.ylim(0.5, 9) # Adjust the limits of the y-axis
plt.xlim(4, 8) # Adjust the limits of the x-axis
sns.set_style("darkgrid")
plt.show()
```

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)

plt.title("Relationship of Sepal and Petal length of Iris Flowers") # Add plot title
plt.ylabel("Petal Length") # Adjust the label of the y-axis
plt.xlabel("Sepal Length") # Adjust the label of the x-axis
plt.ylim(0.5, 9) # Adjust the limits of the y-axis
plt.xlim(4, 8) # Adjust the limits of the x-axis
sns.set_style("ticks") #Execute it twice to see the ticks default format
plt.show()
```
