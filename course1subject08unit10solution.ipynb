```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
```

```python
# Load Iris data

iris = sns.load_dataset("iris")
```

```python
# Violin Plot 

sns.violinplot(x=iris.species, y=iris.petal_length)
plt.title('A violin plot of petal length')
plt.show()
```
