### Solution to Exercise 10

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
```

```python
# Load Iris data

iris = sns.load_dataset("iris")
```

```python
# Violin Plot 

sns.violinplot(x=iris.species, y=iris.petal_length)
plt.title('A violin plot of petal length')
plt.show()
```

### Swamp Plot

```python
sns.swarmplot('species', 'sepal_length', data=iris) # alternative way of passing data
plt.title('A swamp plot of sepal length')
plt.show()
```

```python
sns.swarmplot(x=iris.species, y=iris.sepal_length, data=iris) # alternative way of passing data
plt.title('A swamp plot of sepal length')
plt.show()
```
