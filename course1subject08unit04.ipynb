```python
X = np.linspace(-np.pi, np.pi)
Y = np.sin(X) + np.cos(X)
```

```python
plt.plot(X, Y)
plt.show
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.show
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.show
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.xticks([-np.pi,-np.pi / 2, 0, np.pi / 2, np.pi])
plt.show
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.xticks(
    [-np.pi,-np.pi / 2, 0, np.pi / 2, np.pi], # list of list of xtick locations
    ["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"], # list of labels
)
plt.show()
```
