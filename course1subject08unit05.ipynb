```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-15, 15)
f1 = np.sin(x)/x
f2 = 2*np.sin(x)/x
f3 = 3*np.sin(x)/x

plt.plot(x, f1, color="blue", linestyle= '--', linewidth=2)
plt.plot(x, f2, color="green", linestyle= '-.')
plt.plot(x, f3, color="red", linestyle= 'dotted', marker='o')

plt.show()
```
